```python
class TweetExtractor(object):
    def __init__(self, scraper, dataDir, outPath, metaData, maxNum=None, maxTweetNum=1000):
        self.scraper=scraper
        self.dataDir = dataDir
        self.outPath = outPath
        self.metaData = metaData
        self.maxNum = maxNum
        self.maxTweetNum = maxTweetNum
        
    def primeUSPoliticiansAccounts(self):
        maxNum = self.maxNum
        self.dfPol = pd.read_csv(self.metaData)
        # Only those with twitter accounts
        self.dfPol = dfPol.loc[~dfPol.twitter_account.isnull(), :]
        if maxNum is not None:
            np.random.seed(10)
            men = self.dfPol.loc[self.dfPol.loc[:,"gender"] == 'M',:].sample(int(maxNum/2))
            women = self.dfPol.loc[self.dfPol.loc[:,"gender"] == 'F',:].sample(int(maxNum/2))
            self.metaDatMin = pd.concat([women,men], sort=False)
        self.accts = self.dfPol.loc[:,"twitter_account"] # .dropna()
        
    def getTweetsFromTwitterAccounts(self):
        maxTweetNum = self.maxTweetNum
        self.tweetsDF = pd.DataFrame(# index=range(len(self.accts) * maxTweetNum), 
                                     columns=# ["twitter_account"]+
                                     list(scraper.keys))
        progress = ProgressTracker(len(self.accts), 1)
        dfIndex=0
        for i, acct in enumerate(self.accts):
            filepath = self.dataDir + acct + ".json"
            try:
                with codecs.open(filepath, 'r', encoding='utf8') as rf: 
                    read_tweets = json.load(rf)[:maxTweetNum]
                    read_tweets = [json.loads(t) for t in read_tweets]
            except FileNotFoundError:
                # Explains why groupedDF in IBM-Feed is not size 100; not everyone has collected tweets.
                # print(acct, "not found")
                continue
#             much slower, no idea why: (un-comment setting self.tweetsDF above to see)
#             for tweet in read_tweets:
#                 self.tweetsDF.loc[dfIndex,:] = [acct] + [eval(f"scraper.get{key}")(tweet) for key in self.scraper.keys]
#                 dfIndex += 1
            twoDTweets = [[eval(f"scraper.get{key}")(tweet) for key in self.scraper.keys] for tweet in read_tweets]
            df = pd.DataFrame(twoDTweets)
            df.columns = self.scraper.keys
            df["twitter_account"] = acct
#             self.tweetsDF = pd.concat([self.tweetsDF, df], sort=False)
            self.tweetsDF.append(df, sort=False)
            progress.update(i)
        print("Finished")
        
    def writeOut(self):
        self.dfPol.to_csv(f"{self.outPath}metaData{self.maxNum}.csv", index=False)
        self.tweetsDF.to_csv(f"{self.outPath}tweets{self.maxNum}by{self.maxTweetNum}.csv", index=False)
```
