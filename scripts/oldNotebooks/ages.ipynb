```python
legs2 = pd.read_csv('more-data/legs_dates.csv', encoding='utf-8')
```

```python
legs2
```

```python
dates = [datetime.strptime(i, "%m/%d/%Y") for i in legs2['birthday']]
```

```python
dates[-1]
```

```python
def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
calculate_age(dates[-1])
```

```python
ages = [calculate_age(date) for date in dates]
len(ages)
```
