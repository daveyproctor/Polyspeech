```python
num = 50
pol_header = ['first_name', 'last_name', 'gender', 'age', 'type', 'senate_class', 'party', 'twitter']
tweet_header = ['tweet_id', 'tweet_text', 'tweet_retweet_count', 'tweet_favorite_count']
with open('pol_tweets_{}.csv'.format(num), 'wb') as file:
    writer = unicodecsv.writer(file, delimiter = ',', quotechar = '"')
    writer.writerow(pol_header + tweet_header)
    for _, user in users.iterrows():
        if type(user['twitter']) == float:
            print(user['first_name'], user['last_name'], 'has no twitter')
            continue
        
        user_info = [user[head] for head in pol_header]

        # Get {num} most recent **FULLTEXT, 280 char** tweets for the current user.
        for tweet in Cursor(api.user_timeline, screen_name = user['twitter'], tweet_mode='extended').items(num):
            # Get info specific to the current tweet of the current user.
            tweet_info = [tweet.id, 
                          unidecode(tweet.full_text).replace('\n', '--'), #.replace('.', '--').replace('!', '--').replace('?', '--'),
                          tweet.retweet_count,
                          tweet.favorite_count]

            # Write data to CSV.
            writer.writerow(user_info + tweet_info)

        # Show progress.
        print("Wrote tweets by {} {} username {} to CSV.".format(user['first_name'], user['last_name'], user['twitter']))
```
